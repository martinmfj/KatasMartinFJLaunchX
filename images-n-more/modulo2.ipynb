de desarrollo

TRABJAR CON PAQUETES

Los paquetes son modulos externos o bibliotecas que se incluyen en el proyecto. Como cualquier proyecto es importante conocer como
se va asegurar la presencia de los recursos adecuados para el programa.
Una buenas forma de afrontar esto abordar el programa como un proyecto. Pytthon aborda esto mediante el uso de entornos virtuales.

ENTORNOS VIRTUALES

La ejecucion de programas de forma local en otros dispositivos puede suponer problemas por la compatibilidad de versiones o recursos
disponibles. La solucion es la ejecucion del programa de forma aislada.

Python ofrece un entorno virtual. Un entorno virtual es un copia autonoma de todo lo necesario para ejecutar el programa. Mediante 
un entorno virtual se puede asegurar que el programa contara tendrá acceso a las versiones y recursos correctos

Flujo de trabajo
- Crear el entorno virtual que no afecte el resto de la maquina.
- Ingresar al entorno virtual especificando las versiones y bibliotecas implicadas.
- Desarrollar el programa.

Otras maneras de instalar paquetes

Teniendo un conjunto de archivos en tu máquina e instalándolos desde esa fuente:

Instalar desde un repositorio de GitHub que nos proporciona el control de versiones:
git+https://github.com/your-repo.git

Instalar desde un archivo comprimido:
python3 -m pip install package.tar.gz
